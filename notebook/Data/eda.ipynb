```python
from sklearn.preprocessing import LabelEncoder

columns_to_encode = ['accepts_mercadopago', 'boosted', 'free_shipping', 'fulfillment', 'health', 'is_pdp', 'listing_type_id', 'logistic_type', 'offset', 'original_price', 'price']

le = LabelEncoder()

# Apply LabelEncoder to each column
train_df[columns_to_encode] = train_df[columns_to_encode].apply(lambda col: le.fit_transform(col))
```

```python
train_df=train_df.drop(['category_id','date','deal_print_id','domain_id','etl_version','full_name','main_picture','print_server_timestamp','site_id','sold_quantity','tags','uid','warranty'],axis=1)
```

```python
train_df=train_df.drop(['item_id'],axis=1)
```

```python
train_df=train_df.drop(['platform'],axis=1)
```

```python
train_df=train_df.drop(['title'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_df, train_tgt, test_size=0.33, random_state=42)
```
